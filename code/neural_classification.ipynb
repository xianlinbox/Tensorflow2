```python
from sklearn.datasets import make_circles

n_sample = 1000
x,y = make_circles(n_samples=n_sample, factor=.5, noise=.05, random_state=42)

x,y
```

```python
import pandas as pd

circles = pd.DataFrame({"X0":x[:,0], "X1":x[:,1], "label":y})

circles
```

```python
import matplotlib.pyplot as plt


plt.scatter(x[:,0], x[:,1], c=y, cmap="RdYlBu", s=40)
```

```python
import tensorflow as tf

model_1 = tf.keras.Sequential([
    tf.keras.layers.Dense(units=100, ),
    tf.keras.layers.Dense(units=10),
    tf.keras.layers.Dense(units=1),
])

model_1.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])

model_1.fit(x, y, epochs=100)
```

```python
model_1.evaluate(x,y)
```

```python
model_2 = tf.keras.Sequential([
    tf.keras.layers.Dense(units=4, activation="relu"),
    tf.keras.layers.Dense(units=4, activation="relu"),
    tf.keras.layers.Dense(units=1, activation="sigmoid"),
])

model_1.compile(optimizer=tf.keras.optimizers.Adam(lr=0.001), loss="binary_crossentropy", metrics=["accuracy"])

model_1.fit(x, y, epochs=100)
```
